```python
import xarray as xr
import rioxarray
import zarr
from numcodecs import Blosc
from datetime import datetime
```

```python
ds = xr.open_mfdataset('../../heat_data/GOES_DMV_processed/*', concat_dim='datetime', combine='nested')
```

```python
ds
```

```python
compressor = Blosc(cname='zstd', clevel=3)
encoding = {vname: {'compressor': compressor} for vname in ds.data_vars}
# Save to zarr
save_path = '/Users/jonstar/Documents/heat_data/GOES_DMV_processed.zarr'
ds.to_zarr(store=save_path, encoding=encoding, consolidated=True, mode='w')
```

```python
xr.open_zarr(save_path)
```
