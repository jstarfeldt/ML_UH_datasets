```python
city = cities[23]
print(city)
processed_Landsat_dir = f'{root_directory}/{city}/processed_Landsat'
processed_Landsat_list = sorted(glob.glob(f'{processed_Landsat_dir}/*'))
for file in processed_Landsat_list:
    if 'Landsat9' not in file:
        ds = reprocess_file(file)
        split = file.split('Landsat_')
        fname = split[0] + 'Landsat8_' + split[1]
        ds.to_netcdf(fname)
```
