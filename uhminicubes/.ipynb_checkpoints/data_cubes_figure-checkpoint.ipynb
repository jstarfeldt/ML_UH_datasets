```python
import numpy as np
import matplotlib.pyplot as plt
import svgutils.compose as sc
from IPython.display import SVG
```

```python
# Directory where individual data cube plots are stored
figsave_path = '/glade/derecho/scratch/jstarfeldt/data_analysis/plots'
```

```python
# Visualize the figure to be created
sc.Figure("8in", "4in", \
        # Create a panel for the Landsat/Sentinel-1 data cube, scale it,
        # and add text denoting it as sub-figure a
        sc.Panel(\
              sc.SVG(f'{figsave_path}/Landsat_file_structure.svg').scale(0.01),\
              sc.Text("a)", 20, 18, size=16).scale(0.01)\
             ),\
        # Create a panel for the GOES/MW LST data cube, scale it,
        # and add text denoting it as sub-figure b
        sc.Panel(\
              sc.SVG(f'{figsave_path}/GOES_file_structure.svg').scale(0.01).move(3, 0),\
              sc.Text("b)", 326, 18, size=16).scale(0.01)\
             ))
```

```python
# Creates the identical figure as the previous cell, but saves it as an SVG
sc.Figure("8in", "4in", \
        sc.Panel(\
              sc.SVG(f'{figsave_path}/Landsat_file_structure.svg').scale(0.01),\
              sc.Text("a)", 20, 18, size=16).scale(0.01)\
             ),\
        sc.Panel(\
              sc.SVG(f'{figsave_path}/GOES_file_structure.svg').scale(0.01).move(3, 0),\
              sc.Text("b)", 326, 18, size=16).scale(0.01)\
             )).save(f'{figsave_path}/data_structure_figure.svg')
```
